```python
import time
import random

df_column = df['equipament'].unique()
def generation(df, qtd_samples):
    new_df = []
    for row in df:
        samples = []
        while len(samples) < qtd_samples:
            retry = True
            while retry:
                try:
                    response = chatgpt(f"simule ser um especialista em equipamentos mecânicos e elétricos que trabalha com classificação e categorização de equipamentos e que teria conhecimento sobre os diferentes tipos de equipamentos e seria capaz de fornecer uma lista dos equipamentos específicos para cada grupo de equipamento genérico. agora me faça uma lista python com todos os exemplos que conseguir de equipamentos relacionados a esse grupo: {row}")
                except:
                    time.sleep(60)
                    response = chatgpt(f"simule ser um especialista em equipamentos mecânicos e elétricos que trabalha com classificação e categorização de equipamentos e que teria conhecimento sobre os diferentes tipos de equipamentos e seria capaz de fornecer uma lista dos equipamentos específicos para cada grupo de equipamento genérico. agora me faça uma lista python com todos os exemplos que conseguir de equipamentos relacionados a esse grupo: {row}")
                try:
                    samples += eval(response)
                    retry = False
                except:
                    continue
        if len(samples) > qtd_samples:
            samples = random.sample(samples, qtd_samples)
        dict_ = {'label':row, 'keywords':samples}
        new_df.append(dict_)
        print(row)
    return pd.DataFrame(new_df)
```
